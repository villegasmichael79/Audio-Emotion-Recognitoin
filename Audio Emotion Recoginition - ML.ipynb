```python
import numpy as np
import librosa
import pickle
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler

#%%

class Train:
    def __init__(self, X, y):
        self.X = X
        self.y = y

    def extract_feature(self, X, sample_rate=44000):
        stft = np.abs(librosa.stft(X))
        mfccs = np.mean(librosa.feature.mfcc(y=X, sr=sample_rate, n_mfcc=40).T, axis=0)
        chroma = np.mean(librosa.feature.chroma_stft(S=stft, sr=sample_rate).T, axis=0)
        mel = np.mean(librosa.feature.melspectrogram(X, sr=sample_rate).T, axis=0)
        result = np.hstack((mfccs, chroma, mel))
        return result

    def train(self):
        features = []
        print("Extracting Features...")
        for x in self.X:
            features.append(self.extract_feature(x))
        scaler = StandardScaler()
        features = scaler.fit_transform(features)
        print("Creating Model")
        neigh = KNeighborsClassifier(n_neighbors=1)
        neigh.fit(features, self.y)
        print("Saving Model")
        file = open("knn_Model.pkl", "wb")
        pickle.dump(neigh, file)
        file.close()
        print("Model Saved, Training Completed")

#%%

# Training Set
#path = 'C:/Users/catia/Dropbox (UFL)/Teaching/2020 Fall/EEE 4773 Fundamentals of Machine Learning/GitHub/Final-Project/Code and Reports/Dataset/'
path = 'C:/Users/17869/Downloads/' 
X_train = np.load(path+'data_training.npy')
y_train = np.load(path+'labels_training.npy')


tr = Train(X_train, y_train)
tr.train()

```
